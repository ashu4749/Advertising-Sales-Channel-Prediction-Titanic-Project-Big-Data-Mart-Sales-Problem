# Distribution plot

```python
#load the data
tips=sns.load_dataset("tips")
print(tips.head())
#create violinplot
sns.violinplot(x="total_bill",data=tips)
#show the plot
plt.show()
```

```python
sns.histplot(data=tips,x='total_bill',bins=20)
```

```python
sns.distplot(tips["tip"],bins=10)
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],bins=100)
```

```python
sns.distplot(tips["total_bill"],bins=50,kde=False)
```

# kdeplot

```python
sns.kdeplot(tips["total_bill"])
```

```python
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker")
```

# comosite views onto multivariate datasets

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
sns.rugplot(data=tips,x="tip")
```

```python
sns.rugplot(data=tips,x='total_bill',y='tip')
```

# box plot

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=tips, palette="coolwarm")
```

```python
sns.pairplot(tips)
```
